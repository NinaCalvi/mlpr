```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 10**(6)
vals = np.random.randn(N)
```

```python
plt.figure(figsize=(16,9))
hist = plt.hist(vals, bins=100)
plt.show()
```

Plotting a theoretical prediction of the above histogram

```python
plt.figure(figsize=(16,9))
# hist = (values of bins, edges of the bins, patches)
bin_centres  = 0.5*(hist[1][1:] + hist[1][:-1]) #find center of each bin therefore average the edges of each bin
# Fill in an expression to evaluate the PDF at the bin_centres.
pdf = 1/(2*np.pi)*np.exp(-0.5*(bin_centres**2))
bin_width = bin_centres[1]-bin_centres[0]
predicted_bin_heights = pdf*N*bin_width
plt.plot(bin_centres, predicted_bin_heights, '-r')
plt.show()
```

### Limitation of least square

```python
import numpy as np
import matplotlib.pyplot as plt

# Train model on synthetic dataset
N = 200
X = np.random.rand(N, 1)*10
yy = (X > 1) & (X < 3)
def phi_fn(X):
    return np.concatenate([np.ones((X.shape[0],1)), X, X**2], axis=1)
ww = np.linalg.lstsq(phi_fn(X), yy)[0]

# Predictions
x_grid = np.arange(0, 10, 0.05)[:,None]
f_grid = np.dot(phi_fn(x_grid), ww)

# Predictions with alternative weights:
w2 = [-1, 2, -0.5] # Values set by hand
f2_grid = np.dot(phi_fn(x_grid), w2)

# Show demo
plt.clf()
plt.plot(X[yy==1], yy[yy==1], 'r+')
plt.plot(X[yy==0], yy[yy==0], 'bo')
plt.plot(x_grid, f_grid, 'm-')
plt.plot(x_grid, f2_grid, 'k-')
# limit the axis
plt.ylim([-0.1, 1.1])
plt.show()
```

```python
def Gauss_Class(xx,mu1,mu2,Sigma1,Sigma2,pi1):
    xx = xx.reshape((2,-1))
    Sigma1_I = np.linalg.inv(Sigma1)
    Sigma2_I = np.linalg.inv(Sigma2)
    cmp = np.diag(np.dot((xx-mu1).T,np.dot(Sigma1_I,(xx-mu1)))-np.dot((xx-mu2).T,np.dot(Sigma2_I,(xx-mu2))))
    return cmp>2*np.log(pi1/(1-pi1))

def Gauss_Class_edge(mu1,mu2,Sigma1,Sigma2,offset=10):
    xx = np.random.uniform(-5,15,10000).reshape((2,-1))
    Sigma1_I = np.linalg.inv(Sigma1)
    Sigma2_I = np.linalg.inv(Sigma2)
    cmp = np.diag(np.dot((xx-mu1).T,np.dot(Sigma1_I,(xx-mu1)))-np.dot((xx-mu2).T,np.dot(Sigma2_I,(xx-mu2))))
    xx_p = xx[:,(cmp<offset) & (cmp>0)]
    xx_n = xx[:,(cmp>-offset) & (cmp<0)]
    return xx_p,xx_n
```
