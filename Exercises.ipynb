# Gaussian Stuff

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 10**(6)
vals = np.random.randn(N)
```

```python
plt.figure(figsize=(16,9))
hist = plt.hist(vals, bins=100)
plt.show()
```

Plotting a theoretical prediction of the above histogram

```python
plt.figure(figsize=(16,9))
# hist = (values of bins, edges of the bins, patches)
bin_centres  = 0.5*(hist[1][1:] + hist[1][:-1]) #find center of each bin therefore average the edges of each bin
# Fill in an expression to evaluate the PDF at the bin_centres.
pdf = 1/(2*np.pi)*np.exp(-0.5*(bin_centres**2))
bin_width = bin_centres[1]-bin_centres[0]
predicted_bin_heights = pdf*N*bin_width
plt.plot(bin_centres, predicted_bin_heights, '-r')
plt.show()
```

## Numerical approach integral

```python
mu = 2.0
sigma = 3.0
delta = sigma / 100.0
xx = np.arange(mu - 10*sigma, mu + 10*sigma, delta)
integrand = np.exp(-0.5*(xx - mu)**2/sigma**2)
Zapprox = np.sum(integrand*delta)
print("Zapprox = %g" % Zapprox)
```
